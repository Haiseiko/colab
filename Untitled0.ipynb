```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
import ssl
import urllib3
import sys # Import sys for robust error printing
import os # Added to handle file deletion

# SSL 경고 문구 숨기기 (선택 사항)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

class TLSAdapter(requests.adapters.HTTPAdapter):
    """최신 SSL/TLS 버전을 강제로 사용하도록 설정하는 어댑터"""
    def init_poolmanager(self, *args, **kwargs):
        ctx = ssl.create_default_context(ssl.Purpose.SERVER_AUTH)
        ctx.set_ciphers('DEFAULT@SECLEVEL=1') # 보안 수준 조정 (필요 시)
        ctx.options |= ssl.OP_NO_SSLv3
        ctx.options |= ssl.OP_NO_TLSv1
        ctx.options |= ssl.OP_NO_TLSv1_1
        kwargs['ssl_context'] = ctx
        return super(TLSAdapter, self).init_poolmanager(*args, **kwargs)

def crawl_warrior_seeds():
    print("=== Archive Team Seed Collector (SSL Improved) ===") # Changed to English

    target_input = ""
    while not target_input:
        target_input = input("Target Domain (e.g., dcinside.com): ").strip() # Changed to English
        if not target_input:
            print("❌ Target Domain cannot be empty. Please enter a valid domain.") # Added error message

    try:
        pages_limit = int(input("Max Pages (default 20): ") or 20) # Changed to English
        delay = float(input("Delay (default 1.0): ") or 1.0) # Changed to English
    except:
        pages_limit, delay = 20, 1.0

    print("\n[User-Agent Selection]\n1) PC (Chrome)\n2) Mobile (iPhone)") # Changed to English
    ua_choice = input("Select (default 1): ").strip() # Changed to English

    ua_dict = {
        "1": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
        "2": "Mozilla/5.0 (iPhone; CPU iPhone OS 17_4 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.4 Mobile/15E148 Safari/604.1"
    }

    at_notice = " (ArchiveTeam-URL-Project-Volunteer; gathering seeds for Warrior; See: https://wiki.archiveteam.org/index.php/URLs; This IP is from a public hosting service, not a personal computer. Please do not block.)" # Changed to English
    selected_ua = ua_dict.get(ua_choice, ua_dict["1"]) + at_notice
    headers = {'User-Agent': selected_ua}

    if not target_input.startswith(('http://', 'https://')):
        target_url = "https://" + target_input
    else:
        target_url = target_input

    # Define specific scope for filtering based on the full target_url
    parsed_target_url = urlparse(target_url)
    target_netloc = parsed_target_url.netloc
    target_path_prefix = parsed_target_url.path

    visited, to_visit, all_links = set(), [target_url], set()

    # --- 핵심 개선 부분: 세션 설정 및 SSL 어댑터 연결 ---
    session = requests.Session()
    session.mount("https://", TLSAdapter())
    # ----------------------------------------------

    print(f"\n🚀 Starting crawl: {target_url}\n") # Updated print statement to use target_url

    while to_visit and len(visited) < pages_limit:
        current = to_visit.pop(0)
        if current in visited: continue

        print(f"🔗 [{len(visited)+1}/{pages_limit}] Analyzing: {current}") # Changed to English
        visited.add(current)

        try:
            time.sleep(delay)
            res = session.get(current, headers=headers, timeout=10, verify=True)
            print(f"    HTTP Status: {res.status_code}") # Print HTTP status code

            # 명시적으로 UTF-8로 디코딩하고, 디코딩 오류 발생 시 문자를 대체
            decoded_html = "" # Initialize outside try-except
            try:
                decoded_html = res.content.decode('utf-8', errors='replace')
            except Exception as e_decode:
                # UTF-8 디코딩이 실패하면 requests의 추정 인코딩 또는 latin-1로 대체 시도
                # Print decoding failure message, ensuring UTF-8 output
                print(f"    ⚠️ UTF-8 decoding failed: {e_decode}. Retrying with apparent_encoding.") # Changed to English
                decoded_html = res.content.decode(res.apparent_encoding if res.apparent_encoding else 'latin-1', errors='replace')

            if not decoded_html.strip(): # Check if content is empty or only whitespace
                print(f"    ⚠️ Warning: Decoded HTML content for {current} is empty or invalid. Skipping.") # Changed to English
                continue # Skip this page if content is unusable

            soup = BeautifulSoup(decoded_html, 'html.parser')

            # Extract links from 'a' tags
            for a in soup.find_all('a', href=True):
                full_path = urljoin(current, a['href'])
                # Always add valid HTTP/HTTPS links to all_links, regardless of domain
                if full_path.startswith('http'):
                    all_links.add(full_path)

                    # Add any unvisited HTTP/HTTPS link to to_visit queue
                    if full_path not in visited and full_path not in to_visit:
                        to_visit.append(full_path)

            # Extract links from 'iframe' tags (src attribute)
            for iframe in soup.find_all('iframe', src=True):
                iframe_src = urljoin(current, iframe['src'])
                if iframe_src.startswith('http'):
                    all_links.add(iframe_src)

                    # Add any unvisited HTTP/HTTPS iframe src to to_visit queue
                    if iframe_src not in visited and iframe_src not in to_visit:
                        to_visit.append(iframe_src)

            # Extract links from 'img' tags (src attribute)
            for img in soup.find_all('img', src=True):
                img_src = urljoin(current, img['src'])
                if img_src.startswith('http'):
                    all_links.add(img_src)
                    # Add any unvisited HTTP/HTTPS image src to to_visit queue
                    if img_src not in visited and img_src not in to_visit:
                        to_visit.append(img_src)

        except Exception as e:
            # Robust error printing for outer exception
            error_type = type(e).__name__
            error_message = str(e)
            print(f"    ⚠️ Error ({error_type}): {error_message}") # Changed to English, but error message itself might contain non-ASCII

    # (이후 파일 저장 및 업로드 로직은 이전과 동일)
    if all_links:
        # Sanitize target_url's netloc and path for filename to reflect the specific crawl scope
        # Note: With external crawling, this filename might be less precise in describing content scope.
        full_path_for_filename = urlparse(target_url).netloc + urlparse(target_url).path
        sanitized_filename_base = full_path_for_filename.replace('.', '_').replace('/', '_').replace(':', '_').replace('-', '_')
        # Remove any leading/trailing/multiple underscores that might result from sanitization
        sanitized_filename_base = '_'.join(filter(None, sanitized_filename_base.split('_')))
        filename = f"at_{sanitized_filename_base}.txt"

        try:
            with open(filename, "w", encoding="utf-8") as f:
                f.write("\n".join(sorted(list(all_links))))
            print(f"\n✅ {len(all_links)} links saved to {filename}.") # Changed to English

            # 업로드 시에도 session을 사용하여 SSL 문제 방지
            try:
                with open(filename, 'rb') as f:
                    up_res = session.put(f"https://transfer.archivete.am/{filename}", data=f)
                    if up_res.status_code == 200:
                        print("\n🎉 Upload successful! Enter this address into Warrior:\n" + up_res.text.strip()) # Changed to English
            except Exception as upload_e:
                # Robust error printing for upload failure
                error_type = type(upload_e).__name__
                error_message = str(upload_e)
                print(f"❌ Upload failed: {error_type} - {error_message}") # Changed to English, but error message itself might contain non-ASCII
        finally:
            # Delete the temporary file after processing
            if os.path.exists(filename):
                os.remove(filename)
                print(f"    🗑️ Temporary file '{filename}' deleted.")
    else:
        print("❌ No links found.") # Changed to English

if __name__ == "__main__":
    crawl_warrior_seeds()
```
