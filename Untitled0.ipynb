```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
import ssl
import urllib3
import sys # Import sys for robust error printing
import os # Added to handle file deletion
import concurrent.futures # New import for multi-threading
import threading # New import for thread-safe operations
import queue # New import for a thread-safe queue

# SSL 경고 문구 숨기기 (선택 사항)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

class TLSAdapter(requests.adapters.HTTPAdapter):
    """최신 SSL/TLS 버전을 강제로 사용하도록 설정하는 어댑터"""
    def init_poolmanager(self, *args, **kwargs):
        ctx = ssl.create_default_context(ssl.Purpose.SERVER_AUTH)
        # Removed explicit ciphers and specific TLS version disabling for broader compatibility
        kwargs['ssl_context'] = ctx
        return super(TLSAdapter, self).init_poolmanager(*args, **kwargs)

def crawl_warrior_seeds(target_domain, max_pages, max_workers, delay_seconds, ua_selection, timeout_seconds, crawl_exhaustively=False): # Ensure timeout_seconds is a parameter, Added crawl_exhaustively
    print("=== Archive Team Seed Collector (SSL Improved & Multi-threaded) ===") # Changed title

    target_input = target_domain
    pages_limit = max_pages
    max_workers = max_workers
    delay = delay_seconds
    ua_choice = ua_selection
    request_timeout = timeout_seconds # New: Timeout duration

    if not target_input:
        print("❌ Target Domain cannot be empty. Please enter a valid domain.")
        return # Exit if no target domain

    print("\n[User-Agent Selection]\n1) PC (Chrome)\n2) Mobile (iPhone)")

    ua_dict = {
        "1": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
        "2": "Mozilla/5.0 (iPhone; CPU iPhone OS 17_4 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/17.4 Mobile/15E148 Safari/604.1"
    }

    at_notice = " (ArchiveTeam-URL-Project-Volunteer; gathering seeds for Warrior; See: https://wiki.archiveteam.org/index.php/URLs; This IP is from a public hosting service, not a personal computer. Please do not block.)"
    selected_ua = ua_dict.get(ua_choice, ua_dict["1"]) + at_notice
    headers = {'User-Agent': selected_ua}

    if not target_input.startswith(('http://', 'https://')):
        target_url = "https://" + target_input
    else:
        target_url = target_input

    # Define specific scope for filtering based on the full target_url
    parsed_target_url = urlparse(target_url)
    target_netloc = parsed_target_url.netloc
    target_path_prefix = parsed_target_url.path

    # --- 핵심 개선 부분: 세션 설정 및 SSL 어댑터 연결 --- (Moved inside function and passed to workers)
    session = requests.Session()
    session.mount("https://", TLSAdapter())
    # ----------------------------------------------

    # Thread-safe data structures
    urls_to_visit_queue = queue.Queue() # URLs to be crawled
    visited_urls = set() # All URLs that have been submitted or processed
    all_collected_links = set() # All unique links found across all pages

    # Locks for thread-safe access to shared sets
    visited_lock = threading.Lock()
    all_links_lock = threading.Lock()

    print(f"\n🚀 Starting crawl: {target_url} with {max_workers} threads\n")

    # Seed the queue with the target URL
    urls_to_visit_queue.put(target_url)
    # Removed: visited_urls.add(target_url) # This caused the initial URL to be skipped

    crawled_page_count = 0 # Number of unique pages for which a request was successfully initiated

    # Helper function to fetch and parse a single page
    def _fetch_and_parse_page(url, session_obj, headers_obj, delay_seconds, request_timeout):
        """
        Fetches a single page, parses it, and returns discovered links.
        This function is executed by worker threads.
        """
        discovered_links = set()
        page_status = "N/A"
        print(f"    [WORKER] Fetching {url}...") # Debug print for worker start
        try:
            time.sleep(delay_seconds) # Apply delay per request
            res = session_obj.get(url, headers=headers_obj, timeout=request_timeout, verify=True)
            page_status = res.status_code
            print(f"    [WORKER] Fetched {url} (Status: {page_status}, Content-Length: {len(res.content)})") # More detailed debug print

            decoded_html = ""
            try:
                decoded_html = res.content.decode('utf-8', errors='replace')
            except Exception as e_decode:
                print(f"    ⚠️ UTF-8 decoding failed for {url}: {e_decode}. Retrying with apparent_encoding.")
                decoded_html = res.content.decode(res.apparent_encoding if res.apparent_encoding else 'latin-1', errors='replace')

            if not decoded_html.strip():
                print(f"    ⚠️ Warning: Decoded HTML content for {url} is empty or invalid (Length: {len(decoded_html)}). Skipping {url}.")
                return discovered_links # Return empty set if content is bad
            else:
                print(f"    [WORKER] Decoded HTML content is not empty for {url}. Parsing...")

            soup = BeautifulSoup(decoded_html, 'html.parser')

            # Extract links from 'a', 'iframe', and 'img' tags
            for tag_name, attr_name in [('a', 'href'), ('iframe', 'src'), ('img', 'src')]:
                for tag in soup.find_all(tag_name, **{attr_name: True}):
                    full_path = urljoin(url, tag[attr_name])
                    if full_path.startswith('http'):
                        discovered_links.add(full_path)
            print(f"    [WORKER] Discovered {len(discovered_links)} links from {url} (before filtering).") # Debug print inside worker

        except Exception as e:
            error_type = type(e).__name__
            error_message = str(e)
            print(f"    ⚠️ Error processing {url} ({error_type}): {error_message}")

        return discovered_links # Returns a set of unique links found on this page

    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = {} # Maps Future object to the URL it's processing

        while True:
            # Termination condition 1: If not crawling exhaustively and page limit is reached
            if not crawl_exhaustively and crawled_page_count >= pages_limit:
                print(f"Reached pages_limit ({pages_limit}). Finishing current active tasks and stopping new submissions.")
                break

            # Termination condition 2: If no active tasks and queue is empty
            if not futures and urls_to_visit_queue.empty():
                if crawl_exhaustively:
                    print("No more URLs to visit and no active tasks. Exhaustive crawl complete.")
                else:
                    print("No more URLs to visit and no active tasks. Stopping crawl (either limit reached or no new links discovered).")
                break

            # Attempt to submit new tasks to the executor
            # Only submit if queue is not empty, workers available, AND (exhaustive OR within pages_limit)
            while not urls_to_visit_queue.empty() and len(futures) < max_workers and (crawl_exhaustively or crawled_page_count < pages_limit):
                current_url = urls_to_visit_queue.get() # Get a URL from the queue

                with visited_lock:
                    if current_url in visited_urls:
                        urls_to_visit_queue.task_done()
                        continue # Skip if already being processed or already fully processed

                    visited_urls.add(current_url) # Mark as being processed/submitted

                pages_count_display = 'Unlimited' if crawl_exhaustively else pages_limit
                print(f"Submitting [{crawled_page_count+1}/{pages_count_display}] {current_url} for crawling. Active workers: {len(futures)}")
                future = executor.submit(_fetch_and_parse_page, current_url, session, headers, delay, request_timeout)
                futures[future] = current_url # Store the URL with its future
                crawled_page_count += 1 # Increment count of submitted pages
                urls_to_visit_queue.task_done() # Mark task from queue as done

            # If no tasks are currently active after submission attempts, and the queue is empty,
            # this implies we have nothing more to do.
            if not futures and urls_to_visit_queue.empty():
                if crawl_exhaustively:
                    print("No more URLs to visit and no active tasks after submission attempt. Exhaustive crawl complete.")
                else:
                    print("No more URLs to visit and no active tasks after submission attempt. Stopping crawl.")
                break

            # Wait for at least one active future to complete
            done_futures, _ = concurrent.futures.wait(futures, return_when=concurrent.futures.FIRST_COMPLETED)

            for future in done_futures:
                completed_url = futures.pop(future) # Get the URL associated with the completed future

                try:
                    discovered_links_from_page = future.result()
                    print(f"    ✅ Discovered {len(discovered_links_from_page)} links from {completed_url} (after worker return)") # Modified print

                    # Update the global set of all collected links (thread-safe)
                    with all_links_lock:
                        all_collected_links.update(discovered_links_from_page)

                    # Add new, unvisited links to the queue
                    with visited_lock: # Protect visited_urls when checking/adding
                        new_links_queued_this_round = 0
                        for link in discovered_links_from_page:
                            # Only queue if link is not already visited AND (crawl_exhaustively OR within pages_limit)
                            if link not in visited_urls and (crawl_exhaustively or (crawled_page_count + new_links_queued_this_round < pages_limit)):
                                visited_urls.add(link) # Mark as visited/queued
                                urls_to_visit_queue.put(link) # Add to queue for future processing
                                new_links_queued_this_round += 1
                                # Debug print for queuing decision
                                # print(f"        ➕ Queued new link: {link}") # Too verbose for now
                            elif link in visited_urls:
                                # Debug print if link is already visited
                                pass # print(f"        ➖ Skipping visited link: {link}")
                            elif not crawl_exhaustively and (crawled_page_count + new_links_queued_this_round) >= pages_limit:
                                # Debug print if pages_limit is reached
                                pass # print(f"        🛑 Skipping {link} due to pages_limit ({crawled_page_count + new_links_queued_this_round}/{pages_limit})")

                except Exception as e:
                    print(f"Error processing result for {completed_url}: {e}")

        # After the main loop, wait for any remaining active futures to complete gracefully
        print("Waiting for remaining tasks to complete (if any were submitted before termination)...")
        for future in concurrent.futures.as_completed(futures): # This waits for all submitted tasks to finish
            try:
                # Just consume results to ensure no exceptions are left unhandled
                _ = future.result()
            except Exception as e:
                print(f"Error in a remaining future: {e}")

    if all_collected_links:
        # Sanitize target_url's netloc, path, and query for filename to reflect the specific crawl scope
        parsed_target_url_for_filename = urlparse(target_url)
        # Modified: Use only the netloc for the filename base
        full_path_for_filename = parsed_target_url_for_filename.netloc

        # Replace problematic filename characters with underscores
        sanitized_filename_base = full_path_for_filename
        for char in ['.', '/', ':', '-', '?', '&', '=', '\\', '*', '%']:
            sanitized_filename_base = sanitized_filename_base.replace(char, '_')

        # Remove any leading/trailing/multiple underscores that might result from sanitization
        sanitized_filename_base = '_'.join(filter(None, sanitized_filename_base.split('_')))
        filename = f"at_{sanitized_filename_base}.txt"

        # Initialize a set for existing links
        existing_links = set()
        if os.path.exists(filename):
            print(f"📝 Reading existing links from {filename}...")
            try:
                with open(filename, "r", encoding="utf-8") as f:
                    for line in f:
                        existing_links.add(line.strip())
                print(f"    Found {len(existing_links)} existing links.")
            except Exception as e:
                print(f"    ⚠️ Error reading existing file {filename}: {e}")

        # Merge existing links with newly collected links
        final_links_to_write = existing_links.union(all_collected_links)

        # Calculate and print statistics
        num_existing_links = len(existing_links)
        # New unique links are those in all_collected_links that were not already in existing_links
        num_new_unique_links = len(all_collected_links - existing_links)
        num_total_unique_links = len(final_links_to_write)

        print(f"\n📊 Link Statistics:")
        print(f"    - Links already in file: {num_existing_links}")
        print(f"    - New unique links discovered: {num_new_unique_links}")
        print(f"    - Total unique links to save: {num_total_unique_links}")

        try:
            # Write all unique links (existing + new) back to the file
            with open(filename, "w", encoding="utf-8") as f:
                f.write("\n".join(sorted(list(final_links_to_write))))
            print(f"\n✅ {len(final_links_to_write)} unique links (including previous) saved to {filename}.")

            # Upload using the existing session
            try:
                with open(filename, 'rb') as f:
                    up_res = session.put(f"https://transfer.archivete.am/{filename}", data=f)
                    if up_res.status_code == 200:
                        print("\n🎉 Upload successful! Enter this address into Warrior:\n" + up_res.text.strip())
                    else:
                        print(f"❌ Upload failed with status code: {up_res.status_code} - {up_res.text}")
            except Exception as upload_e:
                error_type = type(upload_e).__name__
                error_message = str(upload_e)
                print(f"❌ Upload failed: {error_type} - {error_message}")
        except Exception as file_write_e:
            error_type = type(file_write_e).__name__
            error_message = str(file_write_e)
            print(f"❌ Error writing links to file {filename}: {error_type} - {error_message}")
    else:
        print("❌ No new links found to add or save.")
```

```python
target_domain = "https://alphawiki.org" # 크롤링할 웹사이트 도메인을 입력하세요 (예: "google.com" 또는 "https://www.google.com")
max_pages = 10000 # 수집할 최대 페이지 수
max_workers = 5 # 동시에 작동할 스레드(워커) 수
delay_seconds = 1 # 요청 간 지연 시간 (초)
ua_selection = "1" # 1: PC (Chrome), 2: Mobile (iPhone)
timeout_seconds = 30 # New: 요청 타임아웃 시간 (초)
crawl_exhaustively = True # New: True로 설정하면 max_pages 제한 없이 모든 링크를 크롤링합니다.

crawl_warrior_seeds(target_domain, max_pages, max_workers, delay_seconds, ua_selection, timeout_seconds, crawl_exhaustively)
```
